```python
def multiple_prediction(loaded_model,model_input,total_step):
    
    prediction_solutions = {}
    std,mean=250,500
    t00 = 0
    size= 96
    
    
    model_input = pad_grids(model_input,lam)   #pad
    model_input = ( numerical_solution[t00:t00+1,:,:] - mean ) / std  #scale 
    model_input = model_input.reshape(1,1,size,size,1)        #reshape to 5d tensor
    
    prediction_solutions[0] = loaded_model.predict(model_input) ; 

    for i in range(1,total_step):
        prediction_solutions[i] = loaded_model.predict(prediction_solutions[i-1])
        prediction_solutions[i-1] = ( prediction_solutions[i-1] * std ) + mean
    prediction_solutions[total_step-1] = ( prediction_solutions[total_step-1] * std ) + mean
    
    return prediction_solutions
```

```python
# def multiple_prediction(loaded_model,model_input,total_step):
    
#     prediction_solutions = {}
#     std,mean=250,500
#     t00 = 0
#     size= 96
#     lam  =
    
#     model_input = pad_grids(model_input,lam)   #pad
#     model_input = ( numerical_solution[t00:t00+1,:,:] - mean ) / std  #scale 
#     model_input = model_input.reshape(1,1,size,size,1)        #reshape to 5d tensor
    
#     prediction_solutions[0] = loaded_model.predict(model_input) ; 

#     for i in range(1,total_step):
#         prediction_solutions[i] = loaded_model.predict(prediction_solutions[i-1])
#         prediction_solutions[i-1] = ( prediction_solutions[i-1] * std ) + mean
#     prediction_solutions[total_step-1] = ( prediction_solutions[total_step-1] * std ) + mean
    
#     return prediction_solutions



def preprocess(inp,lam,size):
    t00 = 0
    inp = pad_grids(inp,lam)
    inp = ( inp[[t00],:,:] - 500 ) / 250
    inp = inp.reshape(1,1,size,size,1)
    return inp


def recursive_prediction(model,inp,K):
    '''
    model : Deep learning model
    inp   : Input solution for the model ( standardized solution)
    K     : iterations for the deep learning model
    '''
    
    solutions =np.zeros((1,K,96,96,1))
    solutions[0,[0]] = model.predict(inp)
    for ki in range(1,K): solutions[ki] = model.predict(solutions[ki-1,)
    return solutions
        
        
    
    
    


def iterations_speedup_analysis(model,grid_size=96,P=10,K=10) :
    '''
    grid_size : square grid size 
    P         : steps per deep learning prediction
    K         : total iterations for numerical solution = K * P
    '''   

    log={}
    error ={}

    mean,std=500,250 #standardization params
    bc1,bc2,bc3,bc4,ic0,lam= 600,500,194,248,254,0.27047  # problem params
    t00 = 0

    n = grid_size-2
    grid = generate_grid(n,bc=(bc1,bc2,bc3,bc4),ic = ic0)
    
    #numerical timings
    tic0=time.time() ; solution = solve(grid.copy(),iters=t00+P*K,Lambda=lam,steps=True);toc0=time.time()
    tic1=time.time() ; prediction = recursive_prediction(model,preprocess(solution,lam,grid_size),0);toc1=time.time()
    
    
    return(solution)
    
#     log[0] = {'Solution_T': toc0-tic0 , 
#               'Prediction_T': toc1-tic1,
#               'MAE' : np.mean(mae())}
    print(log)
```

```python
#solve for iters of a grid size = NxN

numerical_solution_time = {}
deeplearning_inference_time ={}
sample_error = {}

for ki in range(1,5):
    
    size = 192 # Square grid size 
    mean = 500 ; std  = 250  # standardization params
    steps_per_prediction = 100   # steps per prediction
    total_predictions_steps = test_sample 
    bc1,bc2,bc3,bc4,ic0,lam= 600,500,194,248,254,0.27047  # problem params
    t00 = 0 #initial time input

    print(f'iters={ t00+step*total_step} - grid size ={size}x{size}:','\r')
    
    #create grid with boundary and initial conditions
    grid = generate_grid( size - 2 , bc =(bc1,bc2,bc3,bc4),ic=ic0)
    
    # Time required for numerical solution
    tic=time.time() ; numerical_solution = solve(grid.copy(), iters = t00+steps_per_prediction*total_predictions_steps ,Lambda= lam ,steps=True) ; toc=time.time()
    
    numerical_solution_time[ki] = (toc-tic)*1000

    model_input = pad_grids(numerical_solution,lam)   #pad
    model_input = ( numerical_solution[t00:t00+1,:,:] - mean ) / std  #scale 
    model_input = model_input.reshape(1,1,size,size,1)        #reshape to 5d tensor

    #time required for deep learning inference
    tic=time.time() ; deeplearning_solution = multiple_prediction(loaded_model,model_input,total_predictions_steps) ; toc=time.time()
        
    deeplearning_inference_time[ki] = (toc-tic)*1000
    
    sample_error[ki]= np.mean ( [mae(numerical_solution[t00+steps_per_prediction*(l+1),1:-1,1:-1],deeplearning_solution[l][0,0,1:-1,1:-1,0])  for l in range(total_predictions_steps)])

    print(f'MAE:{sample_error[ki]}','\n')
    
```
